```python
df.isnull().sum(axis = 0)
```

```python
is_NaN = df.isnull()
row_has_NaN = is_NaN.any(axis=1)
rows_with_NaN = df[row_has_NaN]
print(row_has_NaN)
df_nans = df[row_has_NaN]
df_nans.head()
```

```python
mask = df['candidate'] == 'Other'
df.loc[mask, 'party'] = 'Other'
df[mask].head()
```

```python
mask = df['candidatevotes'].isnull()
df.loc[mask, 'candidatevotes'] = 0
```

```python
df[mask]
```

```python
df.isnull().sum(axis = 0)
```

```python
mask = df['state_po'].isnull()
df.loc[mask, 'state_po'] = 'Unknown'
df.loc[mask, 'FIPS'] = -1
df[mask]
```

```python
df['FIPS'] = df['FIPS'].astype("int")
df['candidatevotes'] = df['candidatevotes'].astype("int")
df['totalvotes'] = df['totalvotes'].astype("int")
print(f'DF has the following data types:\n{df.dtypes}, \n\nand the corresponding NULL values: {df.isnull().sum(axis = 0)}')
```

```python
mask = (df['party'] == 'Other') | (df['candidate'] == 'Other')
print(f'DF records BEFORE drop: {df.shape}')
df.drop(df[mask].index, inplace = True)
print(f'DF records AFTER drop: {df.shape}')
```
